```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data = {
    'Area': [1000, 1500, 1800, 2400, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'Price': [300000, 450000, 500000, 650000, 800000]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['Area', 'Bedrooms']]
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
error = mean_absolute_error(y_test, predictions)
print("Mean Absolute Error:", error)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```

**Deliverable**
This project demonstrates predictive analysis using a machine learning model.

Steps performed:
- Feature selection was done using Area and Bedrooms.
- Data was split into training and testing sets.
- A Linear Regression model was trained.
- Predictions were made on test data.
- Model performance was evaluated using Mean Absolute Error.

Conclusion:
The model is able to predict house prices based on input features, showing how machine learning can be used for predictive analysis.
